# Calculate PPS

```python
def nRMSE(o, s):
    MSE = np.square(np.subtract(o, s)).mean()
    return math.sqrt(MSE) / np.std(o)
```

```python
def pps(x, y):
    x_train, x_test = train_test_split(x, test_size=0.4)
    y_train, y_test = train_test_split(y, test_size=0.4)

    # Cross-validating decision tree regressor
    # with train data and choosing best model
    # parameters = {
    #     'max_depth': range(1, 20)
    # }
    # clf = GridSearchCV(DecisionTreeRegressor(), parameters, n_jobs=1)
    # clf.fit(X=x.to_numpy().reshape(-1, 1), y=y)
    # tree = clf.best_estimator_
    tree = DecisionTreeRegressor()
    tree.fit(x.to_numpy().reshape(-1, 1), y)

    # scores = cross_val_score(
    #     DecisionTreeRegressor(),
    #     x.to_numpy().reshape(-1, 1), y,
    #     cv=7, scoring='neg_mean_absolute_error'
    # )
    # y_predict = scores.mean()

    # predicting with test values
    y_predict = tree.predict(x.to_numpy().reshape(-1, 1))


    # Create predictions from persistance model.
    # Persistance model always assumes that the next value to predict is the same as it is now
    # y_(t+1) = y_t
    #y_naive = x.shift(1)
    #y_naive = y_naive.fillna(0)

    # Create a dummy regressor with different strategies of naively predicting next hour of data
    # with scikits learn dummyRegresor
    # strategies = maean, median, quantile, constant
    dummy = DummyRegressor(strategy='mean')
    dummy.fit(x, y)
    y_naive = dummy.predict(x)

    # calculating normalised RMSE for predicted values and naive values
    nRMSE_model = nRMSE(y, y_predict)
    nRMSE_naive = nRMSE(y, y_naive)
    MAE_model = mean_absolute_error(y, y_predict)
    MAE_naive = mean_absolute_error(y, y_naive)

    # print(MAE_model, MAE_naive)
    # print(nRMSE_model, nRMSE_naive)

    # if MAE_model > MAE_naive:
    #     return 0
    # else:
    #     return 1 - (MAE_model / MAE_naive)

    print(nRMSE_model / nRMSE_naive)

    return [1 - (nRMSE_model / nRMSE_naive), tree]
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import cupy

def pps_random_forest(data, predictor, target):
    if not data.columns[0] == data.columns[1]:
        x_for_naive = data[predictor]
        y_for_naive = data[target]

        """
        Random forest model
        """
        # data = data.fillna(method='ffill')
        
        print(data.columns, data.size)
        for lag in range(-24, 2):
            data[f'lag_{lag}'] = data[target].shift(lag)
    
        data.dropna(inplace=True)
    
        X = data.drop(target, axis=1)  # replace with your target column
        X_cuda = cupy.array(X)
        y = data[target]
        y_cuda = cupy.array(y)
        
        print(X.size, y.size)
        
        X_train, X_test = train_test_split(X_cuda, test_size=0.2, random_state=42, shuffle=False)
        y_train, y_test = train_test_split(y_cuda, test_size=0.2, random_state=42, shuffle=False)
    
        #rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1, max_features='sqrt', max_depth=20)
        rf_model = XGBRegressor(
            n_estimators=50,      # Number of boosting rounds
            max_depth=5,           # Depth of each tree
            learning_rate=0.5,     # Step size shrinkage
            random_state=42,
            device='cuda',
        )
        rf_model.fit(X_train, y_train)
        
        y_pred = rf_model.predict(X_test)
    
        """
        Persistance model
        """
        y_naive = data[target].shift(1).fillna(0)
        # naive_nRMSEs = []
        # for col in X.columns:
        #     naive_nRMSEs.append(nRMSE(y, y_naive[col]))

        # nRMSE_naive = np.mean(np.array(naive_nRMSEs))
        
        
        """
        Quantile model (median = 0.5, maximum = 1, minimum = 0)
        """
        # dummy = DummyRegressor(strategy='quantile', quantile=1)
        # dummy.fit(X_train, y_train)
        # y_naive = dummy.predict(X_test)
        #nRMSE_naive = nRMSE(y_test, y_naive)
        
        """
        Final calculation
        """
        nRMSE_naive = nRMSE(y, y_naive)
        nRMSE_model = nRMSE(cupy.asnumpy(y_test), cupy.asnumpy(y_pred))
        # nRMSE_naive = naive_nRMSEs[0];
        print(nRMSE_naive, nRMSE_model)
        return 1 - (nRMSE_model / nRMSE_naive)
        # return y_naive
    else:
        return 1
```

# PPS matric calculation

```python
def random_forest_matrix_calculation(data):
    pps_matrix = []
    for icname, icdata in data.items():
        pps_row = []
        for jcname, jcdata in data.items():
            #print(f'Now caluclating: {icname},{jcname}')
            pps_row.append(
               pps_random_forest(data.copy()[[icname, jcname]], icname, jcname)
            )
        pps_matrix.append(pps_row)
    
    pd_mat = pd.DataFrame(pps_matrix, columns=['BZ_GSE','V','N','DST','F','SIGMA$Bz','H_COR'])
    pd_mat.index = ['BZ_GSE','V','N','DST','F','SIGMA$Bz','H_COR']
    
    
    htmap = sn.heatmap(pd_mat, annot=True, cmap='Blues')
    htmap.set(title='Naive Model Persistance')
    plt.xlabel('Predictor')
    plt.ylabel('Target')
    fig = htmap.get_figure()
    # fig.savefig('./data/PPS_naivePersistance_DecisionTree_continuous_data.png') 

random_forest_matrix_calculation(data_NM_relevant)
```

## PPS culculation for $B_z$ and $\sigma B_z$

```python
Bz = data_relevant_interpolated['BZ_GSE'].to_frame()
sBz = data_relevant_interpolated['SIGMA$Bz'].to_frame()
DST = data_relevant_interpolated['DST'].to_frame()
```

```python
for lag in range(-5, 2):  # Short lags for Bz
    Bz[f'Bz_lag_{lag}'] = Bz['BZ_GSE'].shift(lag)
for lag in range(-20, -10):  # Longer lags for sigma Bz
    sBz[f'sigma_Bz_lag_{lag}'] = sBz['SIGMA$Bz'].shift(lag)
```

```python
BzsBzDst = pd.concat([Bz, DST], axis=1)
BzsBzDst.columns
```

```python
BzsBzDst.dropna(inplace=True)

X = BzsBzDst.drop('DST', axis=1)
# X_cuda = cupy.array(X)
y = BzsBzDst['DST']
# y_cuda = cupy.array(y)

X_train, X_test = train_test_split(X, test_size=0.3, random_state=42, shuffle=False)
y_train, y_test = train_test_split(y, test_size=0.3, random_state=42, shuffle=False)

# Reduced model complexity
rf_model = XGBRegressor(
    n_estimators=50,
    max_depth=5,            # Increased depth to capture nuance
    learning_rate=0.1,      # Reduced learning rate
    subsample=0.7,
    colsample_bytree=0.7,
    random_state=42,
    # device='cuda',
)
dt_regressor = DecisionTreeRegressor(
    max_depth=1,
    min_samples_split=10,
    min_samples_leaf=20,
    max_features=None,
    criterion='squared_error',
    random_state=42,
)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

# Calculate nRMSE for the model
nRMSE_model = np.sqrt(mean_squared_error(y_test, y_pred)) / y.std()

# Persistence model using lagged DST only
y_naive = BzsBzDst['DST'].shift(1).fillna(0)
nRMSE_naive = np.sqrt(mean_squared_error(y, y_naive)) / y.std()

# PPS calculation
1 - (nRMSE_model / nRMSE_naive), nRMSE_naive, nRMSE_model, nRMSE_model / nRMSE_naive
```
