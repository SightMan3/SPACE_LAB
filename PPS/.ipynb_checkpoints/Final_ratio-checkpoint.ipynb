```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

# Manualny vypocet pomerov

```python
pers_mask = ["N", "V", "Bz", "Bz_sigma", "NM", "NV", "DST"]
PPS_pers = np.abs(np.array(
        [ -4.13817701e-04, -9.10952208e-03,  1.14700502e-03, -2.52452403e-04,
       -1.83445254e-02, -7.88439880e-03,  4.71692062e-01]
))
FFNN_mask = ["Bz", "NM", "Bz_sigma", "N", "V", "NV", "DST"]
PPS_FFNN = np.abs(np.array(
    [-2.20365696e-04, -1.42383923e-03, -1.38260240e-04, -2.89262610e-03,
        6.45383221e-02, -6.04505004e-02, -1.67759861e-01]
))
```

## Usporiadanie oboch poli s shapely hodnotami podla nejakeho vzoru
vzor :
$$
[B_z, NM, \sigma_{Bz}, N, V, NV, DST]
$$

```python
correct_order = [0, 1, 2, 3, 4, 5] # "Bz", "NM", "Bz_sigma", "N", "V", "NV", "DST"
dict1 = {
   "Bz": 0, "NM": 1, "Bz_sigma": 2, "N": 3, "V": 4, "NV": 5, "DST": 6 
}

FFNN_sort = np.zeros(len(PPS_FFNN))
pers_sort = np.zeros(len(PPS_pers))

for o, v in zip(pers_mask, PPS_pers):
    pers_sort[dict1[o]] = v

for o, v in zip(FFNN_mask, PPS_FFNN):
    FFNN_sort[dict1[o]] = v

FFNN_sort, pers_sort
```

```python
ratio = 1 - (pers_sort / FFNN_sort)
```

```python
hm = sn.heatmap(data=np.array([ratio]), 
                xticklabels=FFNN_mask, 
                yticklabels=["DST"],
               annot=True) 
```
